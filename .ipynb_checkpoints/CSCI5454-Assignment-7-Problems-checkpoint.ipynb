## Answer 2 ( Expected Length: 5 Lines)

Each moth is a vertex in a directed graph while the edges connecting vertices represent the younger or older relationship. The direction is from younger to older. 

We could use DFS to find whether a cycle exists in the directed graph: if the cycle exists, it means there will be contradictions among thoses age relationships, it will be "doesn't make sense", else it will be "make sense".

The brief algorithm is as follows, we use a label array to label weather the vertex is in the stack. If it is in the stack, it means there exists cycles.

`
label = [False for x in range(n)]
//the DFS function
def CheckCycle(G,u,label):  
    if u.seen == False:
        u.seen == True
        label[u] == Ture
    for v in Adj[u]:
         if not v.seen and CheckCycle(G,v,label):
             return Ture
         elif label[v]:
             return Ture
     label[v] = False
     return False
`

Apart from DFS recurrsion, other steps cost constant time so the time complexity is $O(|V|+|E|)=O(n+\binom{n}{2})=O(n^2)$(same as DFS using adjacency list)

__(B):__ 

Use Floyd Warshall algorithm, but for each pair of currencies(vertices), let the edge weight to $R_{i,j}$. Also define the path length as the product of the weights instead of weight sum. We want to find cycle with path length over 1.

Then use another matrix to store the trade length between vertices. Let length of each edge be -1 (length from vertex v to v itself be -inf), in the algorithm, update weight product only when product over 1 and length become "longer". After the algorithm ends, check the diagonal of product matrix and trade length matrix.
