## Problem 1: NeuroSAT (15 points)

A binarized neural network has $n$  inputs $x_1, \ldots, x_n$ which can be set to $\pm 1$, and is composed of hidden neurons $N_1, \ldots, N_K$.

Each neuron $N_i$ is connected to previous neurons $N_{1}, \ldots, N_{i-1}$ using edges with weights $w_{i,1}, \ldots, w_{i,i-1}$ and with inputs $x_1, \ldots, x_n$ using edges with weights $v_{i,1}, \ldots, v_{i,n}$. 



The output $z_i$ of the $i^{th}$ neuron is given by  $$ z_i = \sigma\left( b_i + \sum_{j=1}^{i-1} w_{i,j} z_j + \sum_{k=1}^{n} v_{i,j} x_j \right)$$ 


- $z_{j}$ is the output of the $j^{th}$ neuron.
- $w_{i,j}$ is the weight of the connection between $i^{th}$ neuron in $j^{th}$ neuron (note that $j < i$).
- $v_{i,j}$ is the weight of the connection between $i^{th}$ neuron and $j^{th}$ input.
- $b_{i}$ is a constant called the "bias" of the neuron $N_i$.
- $\sigma$ is the step activation function wherein  $\sigma(x) = \begin{cases} +1 & x \geq 0 \\
-1 & x < 0 \\
\end{cases}$.

The last neuron $N_K$ is designated as the output.

Each input to the network is $+1$ or $-1$. 

*NeuroSAT Problem:* Given a network with all the edge weights and biases, we wish to know if it is possible to provide inputs to the network such that the  output of the last neuron is $+1$.

Show that the NeuroSAT problem is NP-complete.

(a) First show that it is in NP.

(b) Next, provide neural networks that can mimic binary AND/OR gates and the unary NOT gate.

(c) Finally, complete the reduction from the problem of deciding if a given Boolean circuit is satisfiable to 
the problem of solving NeuroSAT.

## Problem 2: Vertex Cover Problem as an 0-1 Integer Linear Program (10 points)

Consider an undirected graph $G$ with $n$ nodes and $m$ edges. A vertex cover is a subset $S$ of the
vertices of the graph such that for every edge $e$, at least one of its end points belongs to the cover.
We wish to find a vertex cover that involves the smallest number of vertices.
Formulate a $0-1$ ILP for the minimal vertex cover problem.

You should clearly answer the following questions:
- What are your decision variables?
- What are the constraints for your problem?
- What is the objective and do you maximize it or minimize it?


## Answer (Expected Size: 6 lines)

- What are your decision variables?

  Each vertex $v$ in the graph has defined a corresponding variable $x_v$. If vertex $v$ belongs to the cover, $x_v=1$, else $x_v=0$.
  
  
- What are the constraints for your problem?

  The constraints are $x_u+x_v\geq1, v,u \in V, \forall e=(u,v)\in E$ and $x_v\in\{0,1\}$  
  
  
- What is the objective and do you maximize it or minimize it?

  The objective is to minimize $\sum_{v\in V}^{}{x_v}$, subject to $x_u+x_v\geq1, v,u \in V, \forall (u,v)\in E$ and $x_v\in\{0,1\}$  


## Problem 3: Min Truth 2-SAT is NP Complete (10 points)

Given that the vertex cover problem is NP-Complete: 
- Input: An undirected graph and bound K.
- Output: Is there a vertex cover with at most $K$ vertices in it. 

See previous problem for definition of vertex cover.

Prove that the Min Truth 2-SAT problem is NP Complete.
- Input: A CNF SAT problem with at most 2 variables per clause and bound $K$
- Output: Is there a satisfying truth assignment which assigns at most $K$ variables to true.


## Answer (Expected Size: 10 lines)

__Prove it's NP:__

Given all the variables with at most K variables set to true, we can verify if the output is TRUE in polynomial time. So the problem is NP.

__Reduction:__

In problem 2, we define corresponding variable $x_v$ for each vertex $v$ in a graph $G$ and for every edge ($\forall e = (u, v)\in E$), at least one vertex is in the vertex cover. This is a NP complete problem. 

In the reduction process, first we regard the edge e(u, v) as a clause $(x_u \vee x_v)$. Then the 2-SAT formula is $\bigwedge\limits_{e_{u, v}\in E}^{}{(x_u \vee x_v)}$.  In this case, if the graph $G$ has a vertex cover of at most size K, the 2_SAT problem will have a satisfying truth assignment which assigns at most K variables to true.

__Prove it's NPC:__

From the fact that vertex cover problem is NPC, we can know that Min Truth 2-SAT problem is NP complete.

## Problem 4: ItemChoice Problem (10 points)

Prove that the ItemChoice is NP complete.
- We have items $1, \ldots, N$. 
- Item number $i$ has cost $c_i$ and value $v_i$.
- We wish to select a subset of items whose total cost is less than or equal to our budget $B$ 

The ItemChoice problem asks if a subset of items whose total costs fall below $B$ can be chosen
such that the value of chosen items exceeds $V$.

Inputs to the problem are $N, B, V $, item costs $c_1, \ldots, c_N$, and item values $v_1, \ldots, v_N$.

**Hint:** Use the SUBSET-SUM problem as the starting point.

## Answer (Expected Size: 12 lines)

__Prove subset-sum is NP:__

Subset-sum peoblem is: given a set $S$ of numbers and a target number $t$, we ask whether there is a subset $S'$ of $S$ that add exactly to $t$. For subset-sum problem, we can verify if a given subset add to the target in polynomial time. So subset-sum problem is NP.

__Reduction for subset-sum__:

For each 3-CNF formula $f$, suppose we have $m$ variables and $n$ clauses, then we create two numbers $u, \neg u$ in set S for each variable $x_i$ and two numbers $x, y$, in $S$ for each clause $C_k$. The numbers are in base 10, where each number contains $m+n$ digits and each digit corresponds to either one variable or one clause. Label all the digits as $a_1,\cdots,a_m, b_1,\cdots,b_n$, $a_i$ are for variables and $b_i$ are for clauses.  For __variable__ $x_i$, define integers $u_i$ and $\neg ui$ as follows:  $a_i$ of both $u_i$ and $\neg u_i$ is 1 and for all other digits $a_j , j != i$, are 0;  $b_k$ of $u_i$ is 1 if and only if $x_i$ appears in clause $C_k$;  $b_k$ of $\neg u_i$ is 1 if and only if $\neg x_i$ appears in clause $C_k$. For __clause__ $C_k$, define integers $x_k$ and $y_k$ as follows: All digits $a_i$ of both $x_k$ and $y_k$ are 0. Digit $b_k$ of $x_k$ is 1; digit $b_k$ of $y_k$ is 2. All other digits $b != k$, of both $x_k$ and $y_k$ are 0. Next, let every digit in target $t$ be: digit having the same location with $a_j$ is 1 and digits having the same location with $b_k$ is 4.  If the subset-sum problem can be reached, $f$ is satisfable. If A is a satisfying vector $[x_1,x_2,\cdots]$ that satifies $f$, $S$ could be constructed for subset-sum problem as follows:  If $x_i$ is true, put $v_i$ into set $S$; otherwise, put $\neg v_i$ into $S$. For clause $C_k$, let $r_k$ be the sum in position $b_k$ of all values in $S$ so far. We know $r_k$ must be at least 1 and at most 3.  If $r_k=1$, put both $x_k$ and $y_k$ into $S$;If $r_k=2$, put $y_k$ into $S$; If $r_k=3$, put $x_k$ into $S$. In this case, sum of numbers at position $b_k$ is 4, and total sum is $t$. 

__Prove subset-sum is NPC:__

From the fact that 3-CNF problem is NPC and can be reduced to subset-sum problem in polynomial time, we can know that subset-sum problem is NPC.

__Prove itemchoice is NP:__  
For this problem, given the subset of items and the items' cost and values, we can easily verify if the cost is below $B$ and value exceeds $V$ in polynomial time. So itemchoice problem is NP. 

__Reduction for itemchoice:__
Reduce subset-sum to this problem. Let $c_i=v_i=s_i$, $B=V=t$. Then the answer yes/no to the two problems are the same.  
$\sum_{i\in S}^{}{c_i}\leq B$ means $\sum_{i\in S}^{}{s_i}\leq t$,  
$\sum_{i\in S}^{}{v_i}\geq V$ means $\sum_{i\in S}^{}{s_i}\geq t$,   
then $\sum_{i\in S}^{}{s_i} = t$

__Prove itemchoice is NPC__:
Subset-sum problem is NPC and can be reduced to itemchoice in polynomial time, so itemchoice is NPC.
