## Q1: Bloom Filters (10 points)

> A bloom filter is a fast set data structure that maintains a set $S$ of keys. One can insert keys into the set and test whether a given key $k$ belongs to the set. It may used in applications where the keys are ``complicated'' objects such as TCP packets or images that are expensive to compare with each other.

> The data structure is an array $T$ of Booleans size $m$ with $l$ different hash functions $h_1, \ldots, h_l$.
>> Initially, `T[i] = FALSE` for all `i`.

> If a key $k$ is to be inserted
>> We compute $i_1 = h_1(k), \ldots, i_l = h_l(k)$.
>> We set $T[i_1] = \cdots T[i_l] = \text{TRUE}$.

> __(A)__ Suppose we wish to find out if an element $k$ is a member of the set by checking if
$T[h_1(k)], \ldots, T[h_l(k)]$ are all true. Explain whether this can lead to a *false positive* i.e,
the approach wrongly concludes that $k$ belongs to the set when it was never inserted; or *false negative*
i.e, the approach wrongly concludes that $k$ does not belong to the set when it does.

> __(B)__ Suppose our hash functions are guaranteed to be uniform. I.e, for any randomly chosen
key $k$, for any hash function $h_i$ and cell $j$, 
>  $$ \mathbb{P}( h_i(k) = j)  = \frac{1}{m} $$
> If $n$ keys are chosen at random and inserted into the filter, compute that probability that any given cell $T[j]$ is set to FALSE after this.

> __(C)__ Use the results from previous set to estimate the probabilisty of a false positive. I.e, some $l$ cells
$i_1, i_2, \ldots, i_l$ are simultaneously set to TRUE.


### Answer 1 { Expected Size: 15 lines}
> (A)

> __This can lead to a false positive but not false negative.__

> __For false negative__, if the hash functions could cause collisions, other elements $x_i$ could possibly satisfy $i_1 = h_1(x_1) = h_1(k), \ldots, i_l = h_l(x_l) = h_l(k)$ and $T[i_1] = \cdots T[i_l]$ are set to be TRUE. In this case, we find $T[h_1(k)], \ldots, T[h_l(k)]$ are all TRUE with $k$ never being inserted.

> __For false negative__, only when $T[i_1],\cdots, T[i_l]$ are not all TRUE, $k$ will be considered as not being inserted. However, as long as $k$ has been inserted, $T[i_1],\cdots, T[i_l]$ will be all TRUE. In this case, the situation where $k$ belongs to the set but $T[i_1],\cdots, T[i_l]$ are not all TRUE will be impossible.

## Q3: Prove Universal Hash Function (15 points)

Suppose we are interested in hashing $n$ bit keys into $m$ bit hash values to hash into a table of size
$2^m$. We view our key  as a bit vector of $n$ bits in binary. Eg., for $n = 4$, the key $14 = \left(\begin{array}{c} 1\\ 1\\ 1\\ 0 \end{array} \right)$.

The hash family is defined by random boolean matrices $H$ with $m$ rows and $n$ columns. To compute the hash function, we perform a matrix multiplication. Eg., with $m = 3$ and $n= 4$, we can have a matrix $H$ such as

$$ H = \left[ \begin{array}{cccc} 0 & 1 & 0 & 1 \\
1 & 0 & 0 & 0 \\
1 & 0 & 1 & 1 \\
\end{array} \right]$$.


The value of the hash function $H(14)$ is now obtained by multiplying

$$ \left[ \begin{array}{cccc} 0 & 1 & 0 & 1 \\
1 & 0 & 0 & 0 \\
1 & 0 & 1 & 1 \\
\end{array} \right] \times \left( \begin{array}{c} 
1\\
1\\
1\\
0\\
\end{array} \right) $$

The matrix multiplication is carried out using AND for multiplication and XOR instead of addition.

(A) For a given matrix $H$ and two  keys $x, y$ that differ only in their $i^{th}$ bits, provide a condition for
$Hx = Hy$ holding. (**Hint** It may help to play with examples where you have two numbers $x, y$ that just differ at a particular bit position. Figure out which entries in the matrix are multiplied with these bits that differ).

(B) Prove that the probability that two keys $x, y$ such that $x \not= y$ collide under the random choice of a matrix $x, y$ is at most $\frac{1}{2^m}$.


## Answer 3 (Expected Size: 15 lines)



> (A)

> __condition__: when all entries in the $i^{th}$ column of $H$ are 0, $Hx = Hy$, for a given matrix $H$ and two  keys $x, y$ that differ only in their $i^{th}$ bits

> __reason__: Based on the rule of matrix multiplication and XOR, we can know that the number of "1" appear in the a nx1 matrix after AND operation will influence the XOR result. Even number of "1" in one row will have "0" in the corresponding position in the final multiplication result while odd number of "1"  in one row will have "1" in that position. In this case, "1" appear in the $i^{th}$ column of $H$ will influence the number of "1" appear in one row, which will make the XOR result flip. If we want to make $Hx=Hy$, we must make sure that the number of "1" won't be influenced by the different bit in x and y's $i{th}$ position. In this case, when all the entries in the $i{th}$ column of $H$ are 0, there won't be "1" appear after AND operation and XOR results won't be influenced. 

> (B)

> when $x \not= y$, $x, y$ collide means $Hx = Hy$. 

> If $x$ and $y$ differ in 1 bit(for example $i{th}$ position), drom Problem (A), we can prove that when $Hx=Hy$, in the position where $x$ and $y$ has different bit, all the entries in that corresponding column of $H$ are "0". it means that all m entries in that corresponding column need to be "0" and the probability is Pr(entries in $H$'s $i^{th}$ column are all 0)=$\frac{1}{2^m}$.  

> If $x$ and $y$ have multiple different bits(for example there are k bits different, 2<=k<=n), from $Hx = Hy$ we have if and only if $H (x + y) = 0$. Suppose bits in position $i_1,\cdot,i_k$ are different in $x, y$, then the result of (x+y) = x XOR y only has "1" in the $i_1{th},\cdot,i_k{th}$ positions. Take $j{th}$ row out of $H$ and denote $H_{j,1}, \cdots, H_{j,n}$ as the elements of $j{th}$ row in $H$. After multiply (x+y) to $H$, the $j{th}$ element of result matrix = 0 XOR 0 ...XOR $H_{j,i_1}$ ... XOR $H_{j,i_k}$... = $H_{j,i_1}$ XOR $H_{j,i_2}$ ... XOR $H_{j,i_k}$

> In this case, only when $H_{j,i_1}$ XOR $H_{j,i_2}$ ... XOR $H_{j,i_k}$ = 0 will make $H(x+y)=0$. When there are even number of "1" in $H_{j,i_1}$, $H_{j,i_2}$ ... $H_{j,i_k}$, will make $H_{j,i_1}$ XOR $H_{j,i_2}$ ... XOR $H_{j,i_k}$ = 0, so the probability will be Pr(there are even "1" in k number of $H_{j,i_1}$, $H_{j,i_2}$ ... $H_{j,i_k}$) = $\frac{\sum_{l=0,2,4,\cdots}^{}{k \choose l}}{2^n}$ = $\frac{1}{2^{n-k+1}}$<=$\frac{1}{2}$ (2<=k<=n)

> expand to all m rows, we have $P(H(x+y)=0)$<=$\frac{1}{2^m}$
